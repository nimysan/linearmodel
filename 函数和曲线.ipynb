```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
# sigmoid

x = np.arange(-10, 10, 0.1)
y = []
for t in x:
    y_1 = 1 / (1 + math.exp(-t))
    y.append(y_1)
plt.plot(x, y, label="sigmoid")
plt.xlabel("x")
plt.ylabel("y")
plt.ylim(0, 1)
plt.legend()
plt.show()
```

```python
#sin & cos曲线
x = np.arange(-10, 10, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y1,label="sin")
plt.plot(x,y2,label="cos",linestyle = "--")
plt.xlabel("x")
plt.ylabel("y")
plt.title('sin & cos')
plt.legend()   #打上标签
plt.show()
```

```python
#f(x)=x2+3
x = np.arange(-10, 10, 1)
y = []
for t in x:
    y_1 = t*t  +3;
    y.append(y_1)
plt.plot(x, y, label="f(x)")
plt.xlabel("x")
plt.ylabel("y")
#plt.ylim(0, 1)
plt.legend()
plt.show()
```

```python
def plot(x, y, label="f(x)"):
    plt.plot(x, y, label=label)
    plt.xlabel("x")
    plt.ylabel("y")
    # plt.ylim(0, 1)
    plt.legend()
    plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import leastsq

def Fun(p,x):                        # 定义拟合函数形式
    a1,a2,a3 = p
    return a1*x**2+a2*x+a3
def error (p,x,y):                    # 拟合残差
    return Fun(p,x)-y
def main():
    x = np.linspace(-10,10,100)  # 创建时间序列
    p_value = [-2,5,10] # 原始数据的参数
    noise = np.random.randn(len(x))+2  # 创建随机噪声
    y = Fun(p_value,x)+noise*2 # 加上噪声的序列
    p0 = [0.1,-0.01,100] # 拟合的初始参数设置
    para =leastsq(error, p0, args=(x,y)) # 进行拟合
    y_fitted = Fun (para[0],x) # 画出拟合后的曲线

    plt.figure
    plt.plot(x,y,'r', label = 'Original curve')
    plt.plot(x,y_fitted,'-b', label ='Fitted curve')
    plt.legend()
    plt.show()
    print (para[0])

if __name__=='__main__':
   main()
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
#lanczos曲线
x = np.arange(-6, 6, 0.1)
a1 = 2
a2 = 3
a3 = 5
y1 = np.sinc(x) * np.sinc(x/a1)
y2 = np.sinc(x) * np.sinc(x/a2)
y3 = np.sinc(x) * np.sinc(x/a3)
plt.plot(x,y1,label="a=2")
plt.plot(x,y2,label="a=3",linestyle="--")
#plt.plot(x,y3,label="a=5",linestyle="--")
plt.title('sinc')
plt.legend()   #打上标签
plt.show()
```

```python
#jpeg的quality参数对应DCT变换的量化程度，quality（0-100）值越高，表示量化损失越小，编码质量越好。
import numpy as np
import math
import matplotlib.pyplot as plt
#jpeg编码的qp曲线
x = np.arange(0.1, 100, 0.1)        #从0.1到100，间隔0.1取点
y = []
for t in x:
    if t < 50:
        y_1 = 5000 / t
    else:
        y_1 = 200 - t * 2
    y.append(y_1)
plt.plot(x, y, label="qp-quality")
plt.xlabel("quality")
plt.ylabel("qp")
plt.ylim(0, 500)
plt.legend()    #显示label
plt.show()
```
